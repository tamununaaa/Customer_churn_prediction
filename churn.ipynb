```python
#Function to print unique values of each column of 'object' type
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
            print(f'{column} : {df[column].unique()}')
```

```python
#Calling function for df1 dataframe
print_unique_col_values(df1)
```

```python
#Replace values like 'No internet service' with No
df1.replace('No phone service','No',inplace=True)
df1.replace('No internet service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#Convert Yes and No to 1 and 0, because ML models interprets 0 and 1
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
#Change categorical value for Gender- Female and Male as 1 and 0
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
df1.gender.unique()
```

```python
#Change categorical value for other columns using One Hot Encoding
#eg. Internet Service
#pd.get_dummies(data=df1,columns=['InternetService'])
#it creates 3 columns for Internet Service-
#InternetService_DSL ; InternetService_Fiber optic ; InternetService_No
```

```python
#Apply One Hot Encoding for 3 columns 
#Storing into new dataframe df2
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(3)
```

```python
df2.dtypes
```

```python
#We want to scale columns like 'MonthlyCharges', 'TotalCharges', 'Tenure'
cols_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_scale] = scaler.fit_transform(df2[cols_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
#We split the data into training and testing data
#We define c as all columns except for column 'churn' and y as the column 'churn'
X = df2.drop('Churn', axis='columns')
Y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=5)
```

```python
#we are using 80% data for training
X_train.shape
```

```python
#we are using 20% data for testing
X_test.shape
```

```python
X_train[:10]
```

```python
#Number of columns in the training dataset
len(X_train.columns)
```

```python
#BUILD A MODEL (ANN) USING TENSORFLOW/KERAS
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
Y_pred = []
for element in yp:
    if element > 0.5:
        Y_pred.append(1)
    else:
        Y_pred.append(0)
```

```python
Y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(Y_test,Y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=Y_test,predictions=Y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
